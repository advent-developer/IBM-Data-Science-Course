## 2. SQLite 3 Database 

```python
import sqlite3
import pandas as pd
```

```python
con = sqlite3.connect('Phil_data.db')
```

```python
# since the data is already in pandas data frame no need to read it on csv function.
# store the df_gdp to sqlite database in order to perform data analysis in sql
df_gdp.to_sql("GDP_data", con, if_exists='replace', index = False, method ='multi')
```

```python
#store the df_grdp to sqlite3 database in order to perform data analysis in SQL
df_grdp.to_sql("GRDP_data",con, if_exists='replace', index = False, method = 'multi')
```

```python
#Retrieve the data from sqlite database
query = "SELECT * FROM GDP_data;"
df = pd.read_sql(query,con)
```

```python
#Retrieve the data from sqlite database
query2="SELECT * FROM GRDP_data;"
df2 = pd.read_sql(query2,con)
```

```python
# this code counts the missing value in each column get from Course 7 lab 2

for column in missing_data1.columns.values.tolist():
    print(column)
    print (missing_data1[column].value_counts())
    print("")  
```
